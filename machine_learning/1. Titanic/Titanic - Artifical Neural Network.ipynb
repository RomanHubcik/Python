```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
# load preprocessed train dataset
df_train = pd.read_csv('titanic_train_preprocessed3.csv')

# create matrix of the features (X) and target (y)
X = df_train.drop('Survived',axis=1) # values used for prediction
y = df_train['Survived'] # values to be predicted

# split data from "titanic_train_preprocessed.csv" to training data (75%) and testing data (25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train
```

```python
# X_train = np.asarray(X_train).astype(np.float32) # cause of error: Failed to convert a NumPy array to a Tensor
# y_train = np.asarray(y_train).astype(np.float32) 
# X_test = np.asarray(X_test).astype(np.float32) 
# y_test = np.asarray(y_test).astype(np.float32) 
# X = np.asarray(X).astype(np.float32)
# y = np.asarray(y).astype(np.float32)

model = Sequential()
model.add(Dense(256, input_shape=(19,), activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=40, batch_size=36, validation_split=0.2, verbose=0)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
#y_pred = np.asarray(y_pred).astype(np.int64)
y_pred
```

```python
# # Calculate the accuracy of the model
# from sklearn.metrics import accuracy_score 
# accuracy = accuracy_score(y_test, y_pred)
# print("Accuracy:", accuracy)


# train model on full train set, with 80/20 CV split
training = model.fit(X_train, y_train, epochs=40, batch_size=36, validation_split=0.2, verbose=0)
val_acc = np.mean(training.history['accuracy'])
print("\n%s: %.2f%%" % ('val_acc', val_acc*100))
```

```python
# once happy with the model accuracy, fit it on all data
model.fit(X, y, epochs=40, batch_size=36, validation_split=0.2, verbose=0)
```
